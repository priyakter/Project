```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
image= cv2.imread('/content/001.jpeg',1)
cv2_imshow(image)

```

```python
image.shape
```

```python
type(image)
```

```python
resize_image = cv2.resize(image,[200,300],interpolation=cv2.INTER_NEAREST)
cv2_imshow(resize_image)
```

```python
resize_image1= cv2.resize(image,[200,300],interpolation=cv2.INTER_LINEAR)
cv2_imshow(resize_image1)
```

```python
cropped_image= image[70:100,120:190]
cv2_imshow(cropped_image)
```

```python
rotate_image=cv2.rotate(image,cv2.ROTATE_90_CLOCKWISE)
cv2_imshow(rotate_image)
```

```python
circle_image= cv2.circle(image,(70,70),30,(255,0,0),2)
cv2_imshow(circle_image)
```

```python
font=cv2.FONT_HERSHEY_SIMPLEX
text_image =cv2.putText(image,'CHAMELEON',[100,80],font,0.5,(255,255,255),2)
cv2_imshow(text_image)
```

```python
picture= cv2.imread('/content/001.jpeg',0)
cv2_imshow(picture)
```

```python
equalized_image= cv2.equalizeHist(picture)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image, cmap="gray")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Equalized Image")
plt.imshow(equalized_image, cmap="gray")
plt.axis("off")

plt.show()
```
