```python
w=[40,50.60,70,80,90,100]
calories_bunt=[200,230,190,220,250,270,235]
plt.plot(w)
plt.plot(calories_bunt)
plt.legend(labels=['Calories Bunt','Weight'],loc='lower right')
```

```python
plt.figure(figsize=(15,5))
w=[40,50.60,70,80,90,100]
calories_bunt=[200,230,190,220,250,270,235]
plt.plot(w,'y--')
plt.plot(calories_bunt,'go')
plt.legend(labels=['Calories Bunt','Weight'],loc='lower right')
plt.xticks(ticks=[0,1,2,3,4,5],labels=['A0','A1','A2','A3','A4','A5']);
```

```python
fig, ax= plt.subplots(nrow=2, ncols=2, figsize=(6,6))
w=[40,50.60,70,80,90,100]
calories_bunt=[200,230,190,220,250,270,235]
ax[0,0].plot(calories_bunt,'go')
ax[0,1].plot(w)
ax[0,0].set_title('Calories Bunt')
ax[0,1].set_title('weight')
ax[0,0].set_xticks(ticks=[0,1,2,3,4,5]);

ax[0,1].set_xticks(ticks=[0,1,2,3,4,5]);

ax[0,0].set_xticklabels(labels=['a','b','c','d','e','f']);
ax[0,1].set_xticklabels(labels=['u','v','w','x','y','z']);
```
